```python
import re
#Removes Punctuations
def remove_punctuations(data):
    punct_tag=re.compile(r'[^\w\s]')
    data=punct_tag.sub(r'',data)
    return data

#Removes HTML syntaxes
def remove_html(data):
    html_tag=re.compile(r'<.*?>')
    data=html_tag.sub(r'',data)
    return data

#Removes URL data
def remove_url(data):
    url_clean= re.compile(r"https://\S+|www\.\S+")
    data=url_clean.sub(r'',data)
    return data

#Removes Emojis
def remove_emoji(data):
    emoji_clean= re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    data=emoji_clean.sub(r'',data)
    url_clean= re.compile(r"https://\S+|www\.\S+")
    data=url_clean.sub(r'',data)
    return data


def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

train_df['review_text']=train_df['review_text'].apply(lambda z: remove_punctuations(z))

train_df['review_text']=train_df['review_text'].apply(lambda z: remove_html(z))
train_df['review_text']=train_df['review_text'].apply(lambda z: remove_url(z))
train_df['review_text']=train_df['review_text'].apply(lambda z: remove_emoji(z))
train_df['review_text'] = train_df['review_text'].apply(remove_stopwords)
```

```python
from multiprocessing import Pool
import pandas as pd
import spacy

nlp = spacy.load('en_core_web_sm')

def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_text = ' '.join([token.lemma_ if token.lemma_ != '-PRON-' else token.text for token in doc])
    return lemmatized_text

def parallelize_lemmatization(texts, func, n_cores=4):
    with Pool(n_cores) as pool:
        results = pool.map(func, texts)
    return results

# Assuming train_df contains a column named 'review_text' with text data
train_df['review_text'] = parallelize_lemmatization(train_df['review_text'], lemmatize_text)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.layers import LSTM, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
train_df_subset = train_df_subset.dropna(subset=['review_text'])
```

```python
train_data, test_data = train_test_split(train_df_subset[['review_text', 'rating']], test_size=0.2, random_state=42)
```

```python
max_features = 5000
maxlen = 100
tokenizer = Tokenizer(num_words=max_features, split=' ')
tokenizer.fit_on_texts(train_df_subset['review_text'])

train_seq = tokenizer.texts_to_sequences(train_data['review_text'])
test_seq = tokenizer.texts_to_sequences(test_data['review_text'])
X_train = pad_sequences(train_seq,maxlen=maxlen)
X_test= pad_sequences(test_seq,maxlen=maxlen)
```

```python
embedding_dim = 128
model = Sequential([
    Embedding(input_dim=max_features,output_dim=embedding_dim, input_length=maxlen),
    LSTM(100, dropout=0.2, recurrent_dropout=0.2),
    Dense(6, activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs = 5
Y_train = train_data['rating'].values
model.fit(X_train,Y_train,epochs=epochs,batch_size=64,validation_split=0.2)
Y_test= test_data['rating'].values
Y_pred = model.predict(X_test)
predictions = [int(round(pred.argmax())) for pred in Y_pred]

accuracy = accuracy_score(Y_test, predictions)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
